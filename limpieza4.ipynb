# Importación de datos

```python
import pandas as pd
import numpy as np

# Cargar el archivo Excel
file_path = r'C:\Users\hp\Desktop\GIT HUB\REPOSITORIOS\Base Carabineros 2023\cuadros-estadísticas-policiales_cch_2023.xlsx'

# Leer todas las hojas del archivo Excel en un diccionario de DataFrames
df = pd.read_excel(file_path, sheet_name='4')

# Si quieres ver las primeras filas del archivo cargado
df.head()
```

```python
# Eliminar las columnas 'Unnamed: 2', 'Unnamed: 3' y 'Unnamed: 4'
df = df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1)
# Convertir la fila 0 en los nombres de columna y eliminar la fila 'head'
df.columns = df.iloc[0]
df = df.iloc[1:]
```

```python
df
```

# Nombrar Columnas NaN

```python
# Nombrar columnas sin nombre con los meses del año
dias = ['Lun', 'Mar', 'Mié', 'Jue', 'Vie', 'Sáb', 'Dom']
sin_nombre = [i for i, col in enumerate(df.columns) if str(col).startswith('Unnamed') or col == '' or pd.isna(col)]
for idx, dia in zip(sin_nombre, dias):
    df.columns = [dia if i == idx else col for i, col in enumerate(df.columns)]
```

```python
df
```

## Crear columna Familia y rellenarla con data correspondiente

```python
# Crear la columna 'Familia'
df['Familia'] = ''
#rellenar con NaN los valores vacíos en la columna 'Familia'
df['Familia'] = df['Familia'].replace('', np.nan) #Esto evita errores al usar ffill
```

```python
# Reordenar columnas: 'Código', 'Familia', 'Familia y materia' como las tres primeras
cols = list(df.columns)
for col in ['Código', 'Familia', 'Familia y materia']:
    cols.remove(col)
df = df[['Código', 'Familia', 'Familia y materia'] + cols]
```

```python
# Rellenar <NA> en 'Familia' con el valor de 'Familia y materia' cuando 'Código' es NaN
df.loc[df['Familia'].isna() & df['Código'].isna(), 'Familia'] = df.loc[df['Familia'].isna() & df['Código'].isna(), 'Familia y materia']
```

```python
# Reemplazar las filas vacías en 'Familia' con el valor de la fila anterior
df['Familia'] = df['Familia'].replace('', pd.NA)
df['Familia'] = df['Familia'].fillna(method='ffill') #Foward fill añade el valor de la fila anterior a las filas vacías
```

```python
df
```

### Resguardo división Detención / Denuncia

```python
# Ingresar el valor 'Categoria' en la columna 'Código' en la fila 1
df.at[df.index[0], 'Código'] = 'Categoria'
```

## Limpiar Filas NaN, correspondientes a sumas

```python
# Eliminar filas donde 'Código' es NaN
df = df[df['Código'].notna()]
```

```python
# Eliminar ultimas filas
df = df.iloc[:-4]
```

```python
df
```

## Separar información Denuncias y Detenciones

```python
# Separar columnas según los valores 'Denuncias' y 'Detenciones' en la fila 0
columnas_fijas = ['Código', 'Familia', 'Familia y materia']
denuncias_cols = columnas_fijas +[col for col in df.columns if df.iloc[0][col] == 'Denuncias']
detenciones_cols = columnas_fijas + [col for col in df.columns if df.iloc[0][col] == 'Detenciones']
denuncias_df = df[denuncias_cols]
detenciones_df = df[detenciones_cols]

denuncias_df
```

```python
# Renombrar columnas de mujeres_df con meses completos en español, mayúscula inicial y sin tildes
dias = ['Lun', 'Mar', 'Mié', 'Jue', 'Vie', 'Sáb', 'Dom']
dias_completos = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
detenciones_df = detenciones_df.rename(columns=dict(zip(dias, dias_completos)))
detenciones_df
```

```python
# Renombrar columna 'Familia y materia' como 'Materia' en ambos DataFrames
denuncias_df= denuncias_df.rename(columns={'Familia y materia': 'Materia'})
detenciones_df = detenciones_df.rename(columns={'Familia y materia': 'Materia'})
```

```python
# Eliminar fila donde 'Código' == 'Categoria' en ambos DataFrames
detenciones_df = detenciones_df[detenciones_df['Código'] != 'Categoria']
denuncias_df = denuncias_df[denuncias_df['Código'] != 'Categoria']
denuncias_df
```

# Unir DataFrame en uno

```python

# Crear columna 'Clasificacion' en ambos DataFrames
detenciones_df['Clasificacion'] = 'Detenciones'
denuncias_df['Clasificacion'] = 'Denuncias'

# Combinar ambos DataFrames en uno solo
DenDetDia_df = pd.concat([detenciones_df, denuncias_df], ignore_index=True)

# Ordenar columnas según los valores 'Denuncias' y 'Detenciones' en la fila 0
columnas_fijas = ['Código', 'Familia', 'Materia','Clasificacion']
DenDetDia_df = DenDetDia_df[columnas_fijas + [col for col in DenDetDia_df.columns if col not in columnas_fijas]]
DenDetDia_df
```

# Descargar

```python
# Exportar DenDetMes_df a un archivo CSV en la carpeta CSV_DATA
DenDetDia_df.to_csv(r'C:\Users\hp\Desktop\GIT HUB\REPOSITORIOS\Base Carabineros 2023\CSV_DATA\DenDetDia_df.csv', index=False)
```
