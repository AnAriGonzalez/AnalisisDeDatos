# Importación de datos

```python
import pandas as pd

# Cargar el archivo Excel
file_path = r'C:\Users\hp\Desktop\GIT HUB\REPOSITORIOS\Base Carabineros 2023\cuadros-estadísticas-policiales_cch_2023.xlsx'

# Leer todas las hojas del archivo Excel en un diccionario de DataFrames
df = pd.read_excel(file_path, sheet_name='1')

# Si quieres ver las primeras filas del archivo cargado
df.head()
```

```python
# Convertir la fila 0 en los nombres de columna y eliminar la fila 'head'
df.columns = df.iloc[1]
df = df.iloc[2:]
```

```python
df
```

## Separar Denuncia de Año

```python
# Separar texto y números en la columna 'Denuncia y año' (separados por espacio)
df[['Denuncia', 'Año']] = df['Denuncia y año/3'].str.split(' ', n=1, expand=True)
```

```python
# Modificar nombre de la columna 'Denuncia y año/3' a 'Año'
df = df.rename(columns={'Denuncia': 'Categoria'})
df
```

```python
# Reordenar columnas: 'Categoria', 'Año'
cols = list(df.columns)
for col in ['Categoria', 'Año']:
    cols.remove(col)
df = df[['Categoria', 'Año'] + cols]
```

```python
# Eliminar la columna 'Denuncia y año/3' del DataFrame
df = df.drop(columns=['Denuncia y año/3', 'Total','Variación porcentual/4'])
df
```

```python
# Eliminar filas 12, 23, 24, 25, 26 y 27 (por posición)
df = df.drop([12, 23, 24, 25, 26, 27])
df
```

# Eliminar DF para evitar confusiones

```python
AnioDenDet_df = df
AnioDenDet_df
del df
```

# Descargar

```python
# Exportar AnioDenDet_df a un archivo CSV en la carpeta CSV_DATA
AnioDenDet_df.to_csv(r'C:\Users\hp\Desktop\GIT HUB\REPOSITORIOS\Base Carabineros 2023\CSV_DATA\AnioDenDet_df.csv', index=False)
```
