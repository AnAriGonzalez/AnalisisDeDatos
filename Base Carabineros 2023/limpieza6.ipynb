# Importación de datos

```python
import pandas as pd
import numpy as np

# Cargar el archivo Excel
file_path = r'C:\Users\hp\Desktop\GIT HUB\REPOSITORIOS\Base Carabineros 2023\cuadros-estadísticas-policiales_cch_2023.xlsx'

# Leer todas las hojas del archivo Excel en un diccionario de DataFrames
df = pd.read_excel(file_path, sheet_name='6')

# Si quieres ver las primeras filas del archivo cargado
df.head()
```

```python
# Eliminar las columnas 'Unnamed: 2'y 'Unnamed: 3' 
df = df.drop(['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3'], axis=1)

# Convertir la fila 1 en los nombres de columna y eliminar la fila 'head'
df.columns = df.iloc[1]
df = df.iloc[2:]
```

```python
df.head()
```

# Nombrar Columnas NaN

```python
# Nombrar columnas sin nombre con REGIONES
regiones_abr = ['AP', 'TA', 'AN', 'AT', 'CO', 'VA', 'RM', 'LI', 'ML', 'NB', 'BI', 'AR', 'LR','LL', 'AI', 'MA']
sin_nombre = [i for i, col in enumerate(df.columns) if str(col).startswith('Unnamed') or col == '' or pd.isna(col)]
for idx, region in zip(sin_nombre, regiones_abr):
    df.columns = [region if i == idx else col for i, col in enumerate(df.columns)]
```

# Separar por tipo de victima

```python
# Crear la columna 'Victima'
df['Victima'] = ''
#rellenar con NaN los valores vacíos en la columna 'Victima'
df['Victima'] = df['Victima'].replace('', np.nan)
```

```python
df
```

```python

# Reordenar columnas: 'Victima', 'Tipo de víctima, consecuencia/2' como las primeras
cols = list(df.columns)
for col in ['Victima', 'Tipo de víctima, consecuencia/2']:
    cols.remove(col)
df = df[['Victima', 'Tipo de víctima, consecuencia/2'] + cols]
```

```python
df.head()
```

```python
# Rellenar columna 'Victima' con los textos indicados en filas específicas
df.iloc[2, df.columns.get_loc('Victima')] = 'VIF A MUJER'
df.iloc[6, df.columns.get_loc('Victima')] = 'VIF A HOMBRE'
df.iloc[10, df.columns.get_loc('Victima')] = 'VIF A NIÑO'
df.iloc[14, df.columns.get_loc('Victima')] = 'VIF A ADULTO MAYOR'
df.iloc[18, df.columns.get_loc('Victima')] = 'VIOLENCIA INTRAFAMILIAR (VIF)'
df
```

```python

# FF# FFill <NA> en 'Victima' con el valor de arriba, hasta que cambie de valor
df['Victima'] = df['Victima'].fillna(method='ffill')
df
```

### Resguardo división Denuncias / Detencion

```python
# Ingresar el valor 'Categoria' en la columna 'Victima' en la fila 1
df.at[df.index[0], 'Victima'] = 'Categoria'
```

```python
# Eliminar filas donde 'Victima' es igual a 'Tipo de víctima, consecuencia/2'
df = df[df['Victima'] != df['Tipo de víctima, consecuencia/2']]
df
```

```python
# Eliminar ultimas filas
df = df.iloc[:-4]
```

## Separar información 'Denuncias' y 'Detenciones'

```python
# Separar columnas según los valores 'Denuncias' y 'Detenciones' en la fila 0
columnas_fijas = ['Victima','Tipo de víctima, consecuencia/2']
denuncias_cols = columnas_fijas + [col for col in df.columns if df.iloc[0][col] == 'Denuncias']
detenciones_cols = columnas_fijas + [col for col in df.columns if df.iloc[0][col] == 'Detenciones']
denuncias_df = df[denuncias_cols]
detenciones_df = df[detenciones_cols]

denuncias_df.head(3)
```

```python
# Renombrar columnas de Detenciones_df con meses completos en español, mayúscula inicial y sin tildes
regiones_abr = ['AP', 'TA', 'AN', 'AT', 'CO', 'VA', 'RM', 'LI', 'ML', 'NB', 'BI', 'AR', 'LR','LL', 'AI', 'MA']
regiones = ['Arica y Parinacota', 'Tarapacá', 'Antofagasta', 'Atacama', 'Coquimbo', 'Valparaíso', 'Metropolitana', "O'Higgins", 'Maule', 'Ñuble', 'Biobío', 'La Araucanía', 'Los Ríos','Los Lagos', 'Aysén', 'Magallanes']
detenciones_df = detenciones_df.rename(columns=dict(zip(regiones_abr, regiones)))
```

```python
# Renombrar columna 'Tipo de víctima, consecuencia/2' como 'Consecuencia' en ambos DataFrames
denuncias_df = denuncias_df.rename(columns={'Tipo de víctima, consecuencia/2': 'Consecuencia'})
detenciones_df = detenciones_df.rename(columns={'Tipo de víctima, consecuencia/2': 'Consecuencia'})


# Eliminar fila donde 'Victima' == 'Categoria' en ambos DataFrames
detenciones_df = detenciones_df[detenciones_df['Victima'] != 'Categoria']
denuncias_df = denuncias_df[denuncias_df['Victima'] != 'Categoria']
denuncias_df
```

# Unir Df

```python
# Crear columna 'Clasificacion' en ambos DataFrames
detenciones_df['Clasificacion'] = 'Detenciones'
denuncias_df['Clasificacion'] = 'Denuncias'

# Combinar ambos DataFrames en uno solo
ConsecuenciaReg_df = pd.concat([detenciones_df, denuncias_df], ignore_index=True)

# Ordenar columnas según los valores 'Denuncias' y 'Detenciones' en la fila 0
columnas_fijas = ['Victima','Consecuencia','Clasificacion']
ConsecuenciaReg_df = ConsecuenciaReg_df[columnas_fijas + [col for col in ConsecuenciaReg_df.columns if col not in columnas_fijas]]
ConsecuenciaReg_df
```

# Exportar a CSV
