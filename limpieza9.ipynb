# Importación de datos

```python
import pandas as pd
import numpy as np

# Cargar el archivo Excel
file_path = r'C:\Users\hp\Desktop\GIT HUB\REPOSITORIOS\Base Carabineros 2023\cuadros-estadísticas-policiales_cch_2023.xlsx'

# Leer todas las hojas del archivo Excel en un diccionario de DataFrames
df = pd.read_excel(file_path, sheet_name='9')

# Si quieres ver las primeras filas del archivo cargado
df.head()
```

```python
# Eliminar las columnas 'Unnamed: 2', 'Unnamed: 3' y 'Unnamed: 4'
df = df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1)
# Convertir la fila 0 en los nombres de columna y eliminar la fila 'head'
df.columns = df.iloc[0]
df = df.iloc[1:]
```

```python
df
```

# Nombrar Columnas NaN

```python
# Nombrar columnas sin nombre con los meses del año
meses = ['ENE', 'FEB', 'MAR', 'ABR', 'MAY', 'JUN', 'JUL', 'AGO', 'SEP', 'OCT', 'NOV', 'DIC']
sin_nombre = [i for i, col in enumerate(df.columns) if str(col).startswith('Unnamed') or col == '' or pd.isna(col)]
for idx, mes in zip(sin_nombre, meses):
    df.columns = [mes if i == idx else col for i, col in enumerate(df.columns)]
```

```python
df
```

## Crear columna Familia y rellenarla con data correspondiente

```python
# Crear la columna 'Familia'
df['Familia'] = ''

#rellenar con NaN los valores vacíos en la columna 'Familia'
df['Familia'] = df['Familia'].replace('', np.nan) #Esto evita errores al usar ffill
```

```python
# Reordenar columnas: 'Código', 'Familia', 'Familia y materia' como las tres primeras
cols = list(df.columns)
for col in ['Código', 'Familia', 'Familia y materia']:
    cols.remove(col)
df = df[['Código', 'Familia', 'Familia y materia'] + cols]
```

```python
# Rellenar <NA> en 'Familia' con el valor de 'Familia y materia' cuando 'Código' es NaN
df.loc[df['Familia'].isna() & df['Código'].isna(), 'Familia'] = df.loc[df['Familia'].isna() & df['Código'].isna(), 'Familia y materia']
```

```python
# Reemplazar las filas vacías en 'Familia' con el valor de la fila anterior
df['Familia'] = df['Familia'].replace('', pd.NA)
df['Familia'] = df['Familia'].fillna(method='ffill') #Foward fill añade el valor de la fila anterior a las filas vacías
```

```python
df
```

### Resguardo división Hombre / Mujer

```python
# Ingresar el valor 'Sexo' en la columna 'Código' en la fila 1
df.at[df.index[0], 'Código'] = 'Sexo'
```

```python
df
```

## Limpiar Filas NaN, correspondientes a sumas

```python
# Eliminar filas donde 'Código' es NaN
df = df[df['Código'].notna()]
```

```python
df
```

```python
# Eliminar ultimas filas
df = df.iloc[:-4]
```

```python
df
```

## Separar información Hombres y Mujeres

```python
# Separar columnas según los valores 'Hombres' y 'Mujeres' en la fila 0
columnas_fijas = ['Código', 'Familia', 'Familia y materia']
hombres_cols = columnas_fijas + [col for col in df.columns if df.iloc[0][col] == 'Hombres']
mujeres_cols = columnas_fijas + [col for col in df.columns if df.iloc[0][col] == 'Mujeres']
hombres_df = df[hombres_cols]
mujeres_df = df[mujeres_cols]

mujeres_df
```

```python
hombres_df
```

```python
# Renombrar columnas de mujeres_df con meses completos en español, mayúscula inicial y sin tildes
meses_abrev = ['ENE', 'FEB', 'MAR', 'ABR', 'MAY', 'JUN', 'JUL', 'AGO', 'SEP', 'OCT', 'NOV', 'DIC']
meses_completos = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
mujeres_df = mujeres_df.rename(columns=dict(zip(meses_abrev, meses_completos)))
mujeres_df
```

```python
# Renombrar columna 'Familia y materia' como 'Materia' en ambos DataFrames
hombres_df = hombres_df.rename(columns={'Familia y materia': 'Materia'})
mujeres_df = mujeres_df.rename(columns={'Familia y materia': 'Materia'})
hombres_df
```

```python
# Eliminar fila donde 'Código' == 'Sexo' en ambos DataFrames
hombres_df = hombres_df[hombres_df['Código'] != 'Sexo']
mujeres_df = mujeres_df[mujeres_df['Código'] != 'Sexo']
hombres_df
mujeres_df
```

```python
# Crear columna 'Sexo' en ambos DataFrames
hombres_df['Sexo'] = 'Hombres'
mujeres_df['Sexo'] = 'Mujeres'

#Unir ambos DataFrames
DetGenMes_df= pd.concat([hombres_df, mujeres_df], ignore_index=True)

#Nuevas columnas fijas
# Fijar columnas ['Código', 'Familia', 'Materia', 'Sexo'] como las primeras en DetGenMes_df
columnas_fijas = ['Código', 'Familia', 'Materia', 'Sexo']
otras_columnas = [col for col in DetGenMes_df.columns if col not in columnas_fijas]
DetGenMes_df = DetGenMes_df[columnas_fijas + otras_columnas]
DetGenMes_df
```
